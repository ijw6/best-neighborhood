```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

# INTRODUCTION 
Here, we are going to look at which neighborhood in Pittsburgh is the best, and which is the worst using a dataset with where arrest took place. To determine this, we are looking at which neighborhood has the least amount of arrests to conclude if it is the best, and which has the most to conclude it is the worst.

```python
arrests = pd.read_csv("PittArrests.csv")
```

Here, we are getting rid of the excess columns to focus on the one that will includes our metric.

```python
arrests2 = arrests.drop(['PK','CCR','AGE' ,'GENDER','RACE', 'ARRESTTIME',
                        'ARRESTLOCATION', 'OFFENSES','INCIDENTLOCATION','INCIDENTZONE','INCIDENTTRACT',
                        'COUNCIL_DISTRICT', 'PUBLIC_WORKS_DIVISION','X','Y'], axis=1)
arrests2
```

```python
arrcount=arrests2['INCIDENTNEIGHBORHOOD'].value_counts()
arrcount
```

```python
arrcount.head(10).plot(kind='bar',title= "Top 10 neighborhoods with the most arrests")
```

```python
arrcount.tail(10).plot(kind='bar', title = "Top 10 neighborhoods with the least arrests")
```

# Conclusion
Out of the top 10 neighborhoods with the most arrests, the Central Business District has the most arrests, making it the "worst" neighborhood in terms of having the most arrests. Following the Central Business District, South Side Flats and Carrick have the second and third most arrests.

Out of the top 10 neighborhoods with the least arrests, Mt. Oliver Neighborhood has the least arrests, making it the "best" neighborhood. This does not take away from the other neighborhoods on this graph, because they are the top 10 "best" in terms of having the least arrests.
